```python
lon_min = -75
lon_max = 15
lat_min = -30
lat_max = 60
tspan = ("2024-05-01", "2025-07-25")
bbox = (lon_min, lat_min, lon_max, lat_max)
```

```python
results = earthaccess.search_data(
    short_name="PACE_OCI_L3M_CARBON",
    temporal=tspan,
    bounding_box=bbox,
    granule_name="*.DAY.*.0p1deg.*",
)
```

```python
paths = earthaccess.open(results)
```

```python
def time_from_attr(ds):
    """Set the time attribute as a dataset variable

    Args:
        ds: a dataset corresponding to a Level-2 or 3 granule

    Returns:
        the dataset with a scalar "time" coordinate
    """
    datetime = ds.attrs["time_coverage_start"].replace("Z", "")
    ds["time"] = ((), np.datetime64(datetime, "ns"))
    ds = ds.set_coords("time")
    return ds

def trim_number_of_lines(ds):
    ds = ds.isel({"number_of_lines": slice(0, 1709)})
    return ds
```

```python
kwargs = {"combine": "nested", "concat_dim": "time"}
prod = xr.open_mfdataset(paths, preprocess=time_from_attr, **kwargs)
dataset = prod
dataset
```

```python
dataset["carbon_phyto"] = dataset["carbon_phyto"].clip(
    min=dataset["carbon_phyto"].attrs["valid_min"],
    max=dataset["carbon_phyto"].attrs["valid_max"],
)
```

```python
dataset_cphyto_mean = dataset['carbon_phyto'].mean(dim='time')
dataset_cphyto_mean=dataset_cphyto_mean.load()
```

```python
dataset_cphyto_mean = dataset_cphyto_mean.sel(lon=slice(-75,15))
```

```python
dataset_cphyto_mean
```

```python
# plot phytoplankton carbonb
lbs = 12
xyfs = 14
mapn = colors.Normalize(vmin=0, vmax=50)
#colors.LogNorm(vmin=0.001, vmax=0.1)
fig = plt.figure(figsize=(10,6))
axs = fig.add_subplot(1,1,1,projection=ccrs.PlateCarree(), aspect='auto')
im=dataset_cphyto_mean.plot(ax=axs,cmap='jet',norm = mapn,transform=ccrs.PlateCarree())
axs.add_feature(cartopy.feature.LAND)
axs.add_feature(cartopy.feature.COASTLINE)
cbar = im.colorbar
cbar.set_label('')
cbar.ax.tick_params(labelsize=lbs)

axs.set_title('')
axs.set_ylabel('Latitude',fontsize=xyfs)
axs.set_xlabel('Longitude',fontsize=xyfs)
axs.set_xticks(np.arange(-75,15,15), crs=ccrs.PlateCarree())
lon_formatter = cticker.LongitudeFormatter()
axs.xaxis.set_major_formatter(lon_formatter)

axs.set_yticks(np.arange(-90,91,30), crs=ccrs.PlateCarree())
lat_formatter = cticker.LatitudeFormatter()
axs.yaxis.set_major_formatter(lat_formatter)
axs.tick_params(axis='both', which='major', labelsize=lbs)
axs.tick_params(axis='both', which='minor', labelsize=lbs)
plt.show()
```

```python
# interpolate PACE data onto AMT track
dataset_POC_data = dataset_POC['poc']
df_PACE_POC = df.copy()
df_PACE_POC['POC'] = -9999
df_PACE_POC['POC_std'] = -9999

for j in range(len(df)):
    temp = dataset_POC['poc'].sel(lat=df.Latitude[j],lon=df.Longitude[j],method='nearest')
    temp_mean = temp.groupby("time.month").mean()
    temp_std = temp.groupby("time.month").std()
    print(j,temp_mean.sel(month=df.month[j],method='nearest').compute().data)
    df_PACE_POC['POC'][j] = temp_mean.sel(month=df.month[j],method='nearest').data
    df_PACE_POC['POC_std'][j] = temp_std.sel(month=df.month[j],method='nearest').data
```
